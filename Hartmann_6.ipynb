```python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

```python
methods=["uniform","conditionalvariance","k_means","gibbon"]
results={}
batch_size=5
for method in methods:
    results[str(method)]=[]
    for file in glob.glob(f'Hartmann_6/TS_*1000_{method}_Y.npy'):
        results[str(method)].append(np.load(file)[:,0])
    results[str(method)] = np.minimum.accumulate(np.array(results[str(method)]).T)+3.86278

    
for method in methods:

    means = np.mean(results[str(method)],1)
    x=range(len(means))
    stds = np.sqrt(np.var(results[str(method)],1)/len(results[str(method)]))
    
    plt.plot(x,means,label=f"{method}")
    plt.fill_between(x,means+stds,means-stds,alpha=0.2)
    

plt.legend()
plt.yscale("log")
plt.ylabel("Regret")
plt.xlabel("# evaluations (B=5)")
plt.title("S-GP-TS Hartmann_6 1000 features / 25 inducing points")
```
